```python
import modin.pandas as pd

file_path = './raw_data.csv'
chunk_size = 20000  # 每个小CSV文件的行数

def split_csv(file_path, chunk_size):
    # 使用 modin.pandas 读取整个大的 CSV 文件
    df = pd.read_csv(file_path)

    file_index = 1
    num_cols = df.shape[1]

    for i in tqdm.tqdm(range(0, num_cols, chunk_size)):
        # 计算当前分块的起始行和结束行索引
        start_idx = i
        end_idx = min(i + chunk_size, num_cols)
        
        # 获取当前分块的数据
        chunk = df.iloc[:, start_idx:end_idx]
        
        # 生成当前分块的文件名
        file_name = f'./拆分数据/split_{file_index}.csv'
        
        if not os.path.exists(f'./拆分数据/'):
            os.makedirs(f'./拆分数据/')

        # 将当前分块保存为一个独立的 CSV 文件
        chunk.to_csv(file_name)

        file_index += 1

    print(f'CSV文件拆分完成，共生成了 {file_index - 1} 个小的CSV文件。')

split_csv(file_path, chunk_size)
```
